## What is the yield curve and why is it important?

A yield curve is a graphical representation that shows the relationship between the *yield* (i.e., interest rate, which is the money you would get back if you lent your money to the government) on securities and the *maturity* (i.e., duration, or the time you loan your money to the government).

Historically, the yield curve has reflected the market’s sense of the economy. The yield curve is important as it can be used to predict future market conditions. 

## How can the yield curve be used as a predictive tool?

The *shape* of the yield curve changes in accordance with the state of the economy. The slope of the yield curve, which is the difference between the yields on short- and long-term securities, can be seen as a forecaster. 
-  **Normal yield curve** - reflects a higher interest  rate in return for a longer duration and a lower interest rate for a shorter duration
    -  this indicates that the economy is thriving or on-the-rise

-  **Flat yield curve** - there is little difference between short- and long-term interest rates for securities
    -  this could indicate slower economic growth and expectations for future inflation might be falling 
-  **Inverted yield curve** - this occurs (rarely) when short-term interest rates are higher than long term interest rates
    -  can result from the expectation of a decline in inflation and is often seen seen as a warning sign of an incoming recession

Let's take a look at these shapes of the yield curves. Please see the plots below. The plots show four yield curves. The yield curves are labeled according to an average taken over a period of one month, thus the curve labeled 2015-8 reflects interest rates in August 2015 for a variety of maturities, and so on. 

```python
matplotlib.rcParams.update({'font.size': 22})
def plot_many(years_months):
    """
    Plots multiple yield curves in one figure.
    """
    f, ax = plt.subplots(figsize=(9,7))
    def plot(year, month, plotargs):
        # print('{}-{}-01'.format(year, month), '{}-{}-28'.format(year, month))
        ax.plot(
            maturity, 
            df['{}-{}-01'.format(year, month):'{}-{}-28'.format(year, month)].mean(), 
            label="{}-{}".format(year, month),
            marker="x",
            markersize=10,
            lw=2,
            **plotargs
        )
    for ym in years_months:
        plot(str(ym[0]), "{:0>2d}".format(ym[1]), ym[2])
    plt.legend()
    plt.ylabel('Yield (%)')
    plt.xlabel('Maturity (years)')
```

```python
"""
"ls" can be [‘solid’ | ‘dashed’, ‘dashdot’, ‘dotted’ | (offset, on-off-dash-seq) | '-' | '--' | '-.' | ':' |]
"""
years_months = ((2017, 3, {}), (2015, 8, {"ls": "dotted"}), (2007, 4, {"ls": "dashed"}), (2000, 8, {"ls": "dashdot"}))
plot_many(years_months)
plt.savefig('yield_curve_current_Mar19.png', dpi=150)
```

The shapes of the yield curves in the above figure are as follows: 
-  2017-3 is a normal yield curve
    - This solid line is the only curve that changes in the two figures below
-  2015-08 is a normal yield curve
    - This curve is shown as a reference for a *normal* yield curve
-  2007-04 is an inverted yield curve
    -  This is the inverted yield curve that formed prior to the great recession
    -  In December, 2005, the curve appeared normal and started to flatten in January 2006
    -  The curve flattened by ~February 2006 and started to invert in August 2006, 16 months before the recession began
-  2000-08 is an inverted yield curve
    - This is the inverted yield curve that formed prior to the early 2000s recession
    - In January 2000, the yield curve was normal and started flattening out in Feb 2000, and the curve started to invert in July 2000
    - This *inverted* yield curve is from August, 2000, seven months before this recession began

The yield curve to focus on in the above figure and in the following two figures is the solid line. In the above plot, based on the shape of the yield curve, early in 2017, the U.S. economy was doing well.

```python
years_months = ((2018, 4, {}), (2015, 8, {"ls": "dotted"}), (2007, 4, {"ls": "dashed"}), (2000, 8, {"ls": "dashdot"}))
plot_many(years_months)
plt.savefig('yield_curve_current_Mar19.png', dpi=150)
```

Following March 2017, short-term interest rates started increasing, yield curve flattening began and continued throughout 2017. In 2018, the yield curve continued to flatten (see above plot) and by the end of the year, it was nearly flat.

```python
years_months = ((2019, 4, {}), (2015, 8, {"ls": "dotted"}), (2007, 4, {"ls": "dashed"}), (2000, 8, {"ls": "dashdot"}))
plot_many(years_months)
plt.savefig('yield_curve_current_Mar19.png', dpi=150)
```

The shape of the solid yield curve in the above figure is as follows: 
-  2019-04 is a flat yield curve

Let’s compare the previous yield curve to what the current U.S. Treasury yield curve is telling us now. On March 22nd, 2019, the yield curve inverted. A Treasury bill maturing in three months yielded 2.45%, which was 0.02 percentage points more than a Treasury maturing in 10 years.
The inversion lasted a week. On March 29th, it flipped again, and short-term interest rates were lower than long-term interest rates by 0.01%.
In April, 2019, the yield curve has been relatively flat. 

## What does the flattening yield curve imply currently? 

We might be headed towards a recession. Yield curves sometimes flatten on the way to inversions, which tend to signal upcoming recessions, which might happen anywhere from a year to 1 ½ or 2 years from now. The yield curve is only indicative of a recession. It is not definitive. But, a recession is bound to happen at some point in the near future, so it might happen soon. There have been a couple of lengthier alarms in the past (e.g.,  there was a very flat curve in late 1998, but it didn't invert for 2 years, and, the yield curve inverted in late 1966, but a recession didn't hit until the end of 1969).

There may be a gradual slowdown in economic growth in 2019 and 2020. Many of the forces that led to strong growth in 2018 are still in play. But, a recession could happen some time in 2020-2021.
